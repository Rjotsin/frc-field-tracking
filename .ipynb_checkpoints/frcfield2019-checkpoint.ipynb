```python
a = pd.read_sql("""
select 
  case when c.PICKUP = 'Preload from auto' then concat('Level',a.AUTO_LVL,
  case when SUBSTRING(c.PLACE ,1, INSTR(c.PLACE,'left')) is not null then 'Left'
       when SUBSTRING(c.PLACE ,1, INSTR(c.PLACE,'left')) is not null then 'Right'
    else '' end) else c.PICKUP end as pickup
, case when c.PLACE = 'dropped' then c.PICKUP else c.PLACE end as place
, count(*) as total
from init_info i
join cycle c
on i.id = c.id
left join auto a
on i.id = a.id
and CYCLE_NO = 0
where TEAM_NO = 614
  and c.PICKUP <> ''
  and c.PLACE <> ''
group by 1,2
order by 3 desc
""",cnx)

```
